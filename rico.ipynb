```
# importar librerias

import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# grafica imagenes estaticas en el notebook
%matplotlib inline 
```

```
images = []
IMAGE_DIRECTORY='/content/drive/My Drive/melo/'

def get_image(image_path):
  image = cv2.imread(image_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  return image

for file in os.listdir(IMAGE_DIRECTORY):
  if not file.startswith('.'):
    images.append(get_image(os.path.join(IMAGE_DIRECTORY, file)))
    
plt.figure(figsize=(20,10))
for i in range(len(images)):
      plt.subplot(1, len(images), i+1)
      plt.imshow(images[i])
```

```
#Display four images
def display4(a, b, c, d, title1='Original', title2='Editada1', title3='Editada2', title4='Editada3'):
    plt.figure(figsize=(12,4))
    plt.subplot(221), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.imshow(b,cmap='gray'), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(223), plt.imshow(c,cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(224), plt.imshow(d,cmap='gray'), plt.title(title4)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
#leer una imagen
#principle of least astonishment(surprice)
img_BGR=cv2.imread('rub07.jpg', cv2.IMREAD_COLOR)
#b,g,r = cv2.split(img_BGR)
    
b=img_BGR[:,:,0]
g=img_BGR[:,:,1]
r=img_BGR[:,:,2]
    
display4(img_BGR,b,g,r,"BGR","b","g","r")
```

```
# Pasar la imagen a rgb
img_RGB=cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)
r,g,b = cv2.split(img_RGB)

display4(img_RGB, r,g,b, "RGB", "r", "g", "b")
```

```
count = 2
plt.figure(figsize=(12,4))
plt.subplot(2,2,1), plt.imshow(img_RGB/255.0), plt.title("original")
plt.axis("off")

for i, lab in enumerate(["R", "G", "B"]):
    crgb = np.zeros(img_RGB.shape)
    crgb[:,:,i] = img_RGB[:,:,i]
    plt.subplot(2,2,count), plt.imshow(crgb/255.0), plt.title(lab)
    plt.axis("off")
    count += 1
 
plt.show()
```
