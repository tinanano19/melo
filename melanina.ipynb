```
def displayGray4(a, b, c, d, title1='Original', title2='Editada1', title3='Editada2', title4='Editada3'):
    plt.figure(figsize=(14,14))
    plt.subplot(221), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.plot(b, color='gray')
    plt.xlabel('intensidad de iluminacion')
    plt.ylabel('cantidad de pixeles')
    plt.subplot(223), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(224), plt.plot(d, color='gray')
    plt.xlabel('intensidad de iluminacion')
    plt.ylabel('cantidad de pixeles')
    plt.show()
    
def displayGray5(a, b, c, d, title1='Original', title2='Editada1', title3='Editada2', title4='Editada3'):
    plt.figure(figsize=(14,14))
    plt.subplot(221), plt.imshow(a, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.plot(b, color='gray')
    plt.xlabel('intensidad de iluminacion')
    plt.ylabel('cantidad de pixeles')
    plt.subplot(223), plt.imshow(c, cmap='gray'), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.subplot(224), plt.plot(d, color='gray')
    plt.xlabel('intensidad de iluminacion')
    plt.ylabel('cantidad de pixeles')
    plt.show()
    
def displayGray2(a, b, title1='Original', title2='Editada1'):
    plt.figure(figsize=(10,10))
    plt.subplot(221), plt.imshow(a,cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(222), plt.plot(b, color='gray')
    plt.xlabel('intensidad de iluminacion')
    plt.ylabel('cantidad de pixeles')
    plt.show()    
```

```
import numpy as np
import cv2

#similarly, Apply Gamma=0.4
gamma_point_four =  np.array(255*(img/255)**0.1,dtype='uint8')
#histograma
hist = cv2.calcHist([img], [0], None, [256], [0,256])
hist1 = cv2.calcHist([gamma_point_four], [0], None, [256], [0,256])
#Display the images in subplots
displayGray4(img, hist, gamma_point_four, hist1,'original','gamma 0.1')
```
